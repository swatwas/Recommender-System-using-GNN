## What are Graph Neural Networks (GNN)?

#### Graph neural network is a class of artificial neural networks for processing input data structured as graphs.

## Difference between CNN and GNN

#### CNN Process:  Involves sliding a box over pixel grids to extract features, condensing information.
#### GNN Process: Utilizes a node and its neighbors to aggregate information, demonstrating a similarity in extracting features but adapted to graph structures.
#### Both processes involve extracting features from local neighbourhoods, however the structure of the data differs. CNNs extract features from regular grids of pixels while GNNs extract features from irregular graph structures. 
#### A 3x3 box in CNN can be abstracted as a central node surrounded by 8 neighbour nodes in GNN. In CNN, the 3x3 box looks at a grid of 9 pixels at a time - one central pixel and 8 surrounding pixels. A weighted sum of these pixel values is computed. Meanwhile for GNN, it aggregates information from these neighbouring nodes to update the feature representation of the central node.

## More on the GNN Model

#### The GNN model consists of layers where each node aggregates information from its neighbors.
#### Layer Interaction: A target node gathers information from its immediate neighbors (first layer) and further aggregations from subsequent layers.

### 1. Message Passing

#### Message passing is the process where each node in a graph receives information (or messages) from its neighboring nodes. These messages combine the features of the neighbors and, optionally, the edges connecting them. The node gathers these messages to update its own representation. This process allows each node to learn not just from itself but from the structure and features of its local neighborhood.

### 2. Aggregation

#### After receiving messages, each node aggregates (summarizes) the information from its neighbors using operations like sum, mean, max, or more complex attention mechanisms. The purpose of aggregation is to combine the features of all neighbors into a single, unified feature vector for the central node.

### 3. Update Step

#### Once the messages are aggregated, the node updates its feature representation by combining its original features with the aggregated information. This update helps the node capture both its attributes and the context provided by its neighbors.
